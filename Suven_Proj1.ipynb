# 1- Performing Analysis of Meteorological Data

```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('weatherHistory.csv')
df.head()
```

```python
df.shape
```

```python
df1 = df.drop(['Daily Summary','Wind Bearing (degrees)','Summary','Precip Type','Visibility (km)','Pressure (millibars)'],axis=1)
df1.head()
```

```python
df1.shape
```

```python
#df.info
```

```python
#df.isnull()
```

```python
d1=df1[['Formatted','Temperature (C)','Apparent Temperature (C)','Humidity','Wind Speed (km/h)']]
d1
d1.columns = ['Date', 'Temperature', 'ApparentTemperature','Humidity','Wind Speed']
d1 
```

```python
print(d1)
```

```python
d1['Date'] = pd.to_datetime(d1['Date'],utc=True)
d2=d1.set_index('Date')
d2
```

```python
d2.describe()
```

```python
df2 = d2[['ApparentTemperature','Humidity','Temperature']].resample('MS').mean()
df2
```

```python
april = df2[df2.index.month==4]
plt.figure(figsize=(10,5))
plt.plot(april.loc['2006-04-01':'2016-04-01','ApparentTemperature'],marker='*', label = 'Apparent Temperature(C)', color ='green',linestyle='-');
plt.plot(april.loc['2006-04-01':'2016-04-01','Humidity'], marker='*',label = 'Humidity', color  = 'blue');
plt.legend(loc=7, fontsize =12);
plt.xlabel('Month of April',fontsize = 12);
plt.title('Humidity v/s Apparent Temperature(C)', fontsize = 14);
```

```python
plt.figure(figsize=(12,5));
plt.plot(df2['Humidity'], label = 'Humidity', color ='green');
plt.plot(df2['ApparentTemperature'], label = 'A_temp(C).', color  = 'maroon');
plt.title('Variation of Apparent Temperature(C) v/s Humidity', fontsize = 15);
plt.legend(loc=1, fontsize =12);
plt.xticks(fontsize = 15);
plt.yticks(fontsize=13);
```

```python
fig,ax=plt.subplots(figsize=(15,6))
plt.hist(df2["Temperature"],bins=10,rwidth=0.6)
plt.xlabel("Temperature")
plt.ylabel("Frequency")
plt.show()
```
